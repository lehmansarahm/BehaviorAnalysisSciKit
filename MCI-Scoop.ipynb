```python
import pandas as pd
import os
import tarfile
from six.moves import urllib

# some constants ...
DEFAULT_FILENAME = "YA-Single-Breakfast-Scoop-Aggregate.csv"
DEFAULT_DATASET_PATH = "datasets/mci"

# base data load function
def load_mci_file(dataset_path=DEFAULT_DATASET_PATH, filename=DEFAULT_FILENAME):
    csv_path = os.path.join(dataset_path, filename)
    print(csv_path)
    return pd.read_csv(csv_path)

# returns CSV file as a Pandas data frame
```

```python
# call the data load function and display the initial results
mci = load_mci_file()
mci.head()
```

```python
# spits out high level data table info
mci.info()
```

```python
# spits out meta data / statistics about the different table attributes
mci.describe()
```

```python
# Shows histograms of all attributes for this data set
import matplotlib.pyplot as plt

mci.hist(bins=50, figsize=(20,150))
plt.show()
```

```python
# split the dataset into training (80%) and testing (20%) partitions
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(mci, test_size=0.2, random_state=42)
print(len(train_set), "train +", len(test_set), "test")
```

```python
# retrieve the data fields and subsequent labels we want to process
task_labels = train_set["Label"].copy().as_matrix()
task_accel_values = train_set[["Mean-Mag", "Variance-Mag", "Skewness-Mag", "Kurtosis-Mag", "RMS-Mag"]].copy().as_matrix()

# should match
print(len(task_labels), "labels +", len(task_accel_values), "values") 

#print(task_labels.dtype)
#print(task_labels)
#print(task_accel_values.dtype)
#print(task_accel_values)
```

```python
# applying random forest classifier to training set
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=10)
rfc = rfc.fit(task_accel_values, task_labels)
```

```python
# verifying known records ...
# -------------------------------
# source: YA11 end-stir-2, record row: 19, label: non-select
# feature values:  0.811019, 0.003536, 4.007512 e-08, 1.626166 e-10, 0.814178
ya11_end_stir = task_accel_values[0] 
ya11_end_stir
```

```python
# verifying known records ...
# -------------------------------
# source: YA8 select-coffee, record row: 413, label: select
# feature values:  0.833351, 0.016837, 4.949185 e-08, 1.22632 e-10, 0.849058
ya8_select_coffee = task_accel_values[3] 
ya8_select_coffee
```

```python
# attempting to predict labels of known records ...
# should be an array with a single element: zero (false)
# -------------------------------
rfc.predict([ya11_end_stir])
```

```python
# attempting to predict labels of known records ...
# should be an array with a single element: one (true)
# -------------------------------
rfc.predict([ya8_select_coffee])
```

```python
# SUCCESS!!  That's just the training set ...
# Now let's evaluate our performance.

# -------------------------------
#
#        CONFUSION MATRIX
#
# -------------------------------

from sklearn.model_selection import cross_val_predict
label_predictions = cross_val_predict(rfc, task_accel_values, task_labels, cv=3)

from sklearn.metrics import confusion_matrix
confusion_matrix(task_labels, label_predictions) # actual predictions
```

```python
confusion_matrix(task_labels, task_labels) # perfect predictions
```

```python
# ------------------------------------------
#
# PRECISION:
# What percentage of the time is a positive hit truly an accurate hit
#
# ------------------------------------------

from sklearn.metrics import precision_score, recall_score, f1_score
precision_score(task_labels, label_predictions)
```

```python
# ------------------------------------------
#
# RECALL:
# What percentage of true positives are found
#
# ------------------------------------------

recall_score(task_labels, label_predictions)
```

```python
# ------------------------------------------
#
# F-SCORE:
# Harmonic mean of precision and recall
#
# ------------------------------------------

f1_score(task_labels, label_predictions)
```
